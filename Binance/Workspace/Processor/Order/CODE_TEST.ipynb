```python
import asyncio
import nest_asyncio
from typing import List


nest_asyncio.apply()
```

```python


class OrderManager:
    def __init__(self):
        # self.symbol:List = SystemConfig.Streaming.symbols[0]
        self.symbol:List = "TRXUSDT"
        self.margin_type = SystemConfig.Position.margin_type
        # self.leverage = SystemConfig.Position.leverage
        self.leverage = 3
        self.path_api = SystemConfig.Path.bianace
        self.api_key = base_utils.load_json(self.path_api)
        self.ins_tr_client = FuturesTradingClient(**self.api_key)

    async def set_leverage(self):
        """
        레버리지 값을 설정한다. 기존 보유중인 포지션이 있는 상태에서 변경시 오류발생한다.
        """
        await self.ins_tr_client.set_leverage(self.symbol, self.leverage)

    async def set_margin_type(self):
        """
        마진 타입을 설정한다. 기존과 같은 값을 설정시 오류가 발생한다.
        """
        await self.ins_tr_client.set_margin_type(self.symbol, self.margin_type)

    async def check_position(self):
        return await self.ins_tr_client.fetch_account_balance()

dummy_instance = OrderManager()
```

```python
asyncio.run(dummy_instance.check_position())
```
